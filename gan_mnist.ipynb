```python
                                                                                                                     # Google Colab setup for DCGAN on MNIST with GPU T4

import tensorflow as tf
print("TensorFlow version:", tf.__version__)
print("GPUs available:", tf.config.list_physical_devices('GPU'))

from google.colab import drive
drive.mount('/content/drive')  # Follow prompt to authorize

from tensorflow.keras import layers, losses, optimizers, metrics, models
import numpy as np
import matplotlib.pyplot as plt
import os

tf.random.set_seed(42)
np.random.seed(42)

(train_images, _), (_, _) = tf.keras.datasets.mnist.load_data()
train_images = train_images.reshape(-1, 28, 28, 1).astype('float32')
# Normalize to [-1, 1]
train_images = (train_images - 127.5) / 127.5

BUFFER_SIZE = 60000
BATCH_SIZE = 256
train = tf.data.Dataset.from_tensor_slices(train_images)
train = train.shuffle(BUFFER_SIZE).batch(BATCH_SIZE)

# 5. Define Generator architecture
def make_generator_model(latent_dim=100):
    model = tf.keras.Sequential(name='generator')
    model.add(layers.Dense(7*7*256, use_bias=False, input_shape=(latent_dim,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())
    model.add(layers.Reshape((7, 7, 256)))  # 7x7x256

    model.add(layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU())

    model.add(layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh'))
    return model

# 6. Define Discriminator architecture
def make_discriminator_model():
    model = tf.keras.Sequential(name='discriminator')
    model.add(layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same', input_shape=[28, 28, 1]))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'))
    model.add(layers.LeakyReLU())
    model.add(layers.Dropout(0.3))

    model.add(layers.Flatten())
    model.add(layers.Dense(1))  # outputs logits
    return model

class DCGAN(models.Model):
    def __init__(self, discriminator, generator, latent_dim):
        super(DCGAN, self).__init__()
        self.discriminator = discriminator
        self.generator = generator
        self.latent_dim = latent_dim

    def compile(self, d_optimizer, g_optimizer,
                loss_fn=losses.BinaryCrossentropy(from_logits=True)):
        super(DCGAN, self).compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer
        self.loss_fn = loss_fn
        self.d_loss_metric = metrics.Mean(name="d_loss")
        self.g_loss_metric = metrics.Mean(name="g_loss")

    @property
    def metrics(self):
        return [self.d_loss_metric, self.g_loss_metric]

    def train_step(self, real_images):
        batch_size = tf.shape(real_images)[0]
        random_latent_vectors = tf.random.normal(shape=(batch_size, self.latent_dim))

        with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
            # Generate fake images
            generated_images = self.generator(random_latent_vectors, training=True)
            # Discriminator predictions
            real_logits = self.discriminator(real_images, training=True)
            fake_logits = self.discriminator(generated_images, training=True)

            # Label smoothing / noise
            real_labels = tf.ones_like(real_logits) - 0.1 * tf.random.uniform(tf.shape(real_logits))
            fake_labels = tf.zeros_like(fake_logits) + 0.1 * tf.random.uniform(tf.shape(fake_logits))

            # Discriminator loss
            d_loss_real = self.loss_fn(real_labels, real_logits)
            d_loss_fake = self.loss_fn(fake_labels, fake_logits)
            d_loss = 0.5 * (d_loss_real + d_loss_fake)

            # Generator loss (pretend fakes are real)
            g_loss = self.loss_fn(tf.ones_like(fake_logits), fake_logits)

        # Compute gradients
        grads_disc = disc_tape.gradient(d_loss, self.discriminator.trainable_variables)
        grads_gen = gen_tape.gradient(g_loss, self.generator.trainable_variables)
        # Apply gradients
        self.d_optimizer.apply_gradients(zip(grads_disc, self.discriminator.trainable_variables))
        self.g_optimizer.apply_gradients(zip(grads_gen, self.generator.trainable_variables))

        # Update metrics
        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)
        return {"d_loss": self.d_loss_metric.result(), "g_loss": self.g_loss_metric.result()}

latent_dim = 100
generator = make_generator_model(latent_dim)
discriminator = make_discriminator_model()

dcgan = DCGAN(discriminator=discriminator, generator=generator, latent_dim=latent_dim)
dcgan.compile(
    d_optimizer=optimizers.Adam(learning_rate=0.0002, beta_1=0.5),
    g_optimizer=optimizers.Adam(learning_rate=0.0002, beta_1=0.5)
)

class GANMonitor(tf.keras.callbacks.Callback):
    def __init__(self, num_images=5, latent_dim=100):
        self.num_images = num_images
        self.latent_dim = latent_dim
        self.epochs = []
        self.d_losses = []
        self.g_losses = []

    def on_epoch_end(self, epoch, logs=None):
        # Record metrics
        self.epochs.append(epoch + 1)
        self.d_losses.append(logs.get('d_loss'))
        self.g_losses.append(logs.get('g_loss'))

        # Every 10 epochs: display images and loss curves
        if (epoch + 1) % 10 == 0:
            # Generate and display images
            random_latent_vectors = tf.random.normal(shape=(self.num_images, self.latent_dim))
            generated_images = self.model.generator(random_latent_vectors)
            generated_images = (generated_images * 127.5 + 127.5).numpy().astype('uint8')

            fig, axes = plt.subplots(2, self.num_images, figsize=(self.num_images * 2, 4))
            # Plot images
            for i, img in enumerate(generated_images):
                axes[0, i].imshow(img.squeeze(), cmap='gray')
                axes[0, i].axis('off')
            # Plot loss curves
            axes[1, 0].plot(self.epochs, self.d_losses, label='d_loss')
            axes[1, 0].plot(self.epochs, self.g_losses, label='g_loss')
            axes[1, 0].set_xlabel('Epoch')
            axes[1, 0].set_ylabel('Loss')
            axes[1, 0].legend()
            for j in range(1, self.num_images):
                axes[1, j].axis('off')
            plt.show()

EPOCHS = 1000

dcgan.fit(
    train,
    epochs=EPOCHS,
    callbacks=[GANMonitor(num_images=5, latent_dim=latent_dim)]
)
```
