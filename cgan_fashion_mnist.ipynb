```python
# Google Colab setup for CGAN on Fashion-MNIST with GPU T4

# 1. Imports, GPU check, and mount Google Drive
import tensorflow as tf
import os, numpy as np, matplotlib.pyplot as plt
from tensorflow.keras import layers, models, optimizers, metrics
from tensorflow.keras.callbacks import Callback, ReduceLROnPlateau
from google.colab import drive

print("TensorFlow version:", tf.__version__)
print("GPUs available:", tf.config.list_physical_devices('GPU'))

drive.mount('/content/drive', force_remount=True)

# 2. Prepare directory for results in Drive
BASE_DIR = "/content/drive/MyDrive/DCGAN_results"
os.makedirs(BASE_DIR, exist_ok=True)
existing = sorted([d for d in os.listdir(BASE_DIR) if d.startswith("result")])
run_id = len(existing)
RUN_DIR = os.path.join(BASE_DIR, f"result{run_id}")
os.makedirs(RUN_DIR, exist_ok=True)
print("Saving this run to:", RUN_DIR)

# 3. Load and preprocess Fashion-MNIST dataset with labels
from tensorflow.keras.datasets import fashion_mnist
(train_images, train_labels), _ = fashion_mnist.load_data()
train_images = train_images.reshape(-1, 28, 28, 1).astype('float32')
train_images = (train_images - 127.5) / 127.5  # Normalize to [-1, 1]

BUFFER_SIZE = 60000
BATCH_SIZE  = 256
train_ds = (
    tf.data.Dataset
      .from_tensor_slices((train_images, train_labels))
      .shuffle(BUFFER_SIZE)
      .batch(BATCH_SIZE, drop_remainder=True)
      .prefetch(tf.data.AUTOTUNE)
)

# 4. CGAN generator architecture
num_classes = 10
latent_dim = 100

def make_generator_model(latent_dim=100, num_classes=10):
    noise_input = layers.Input(shape=(latent_dim,))
    label_input = layers.Input(shape=(1,), dtype='int32')
    label_embedding = layers.Embedding(num_classes, latent_dim, input_length=1)(label_input)
    label_embedding = layers.Flatten()(label_embedding)
    x = layers.Concatenate()([noise_input, label_embedding])
    x = layers.Dense(7*7*256, use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Reshape((7, 7, 256))(x)
    x = layers.Conv2DTranspose(128, 5, strides=1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Conv2DTranspose(64, 5, strides=2, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)
    output = layers.Conv2DTranspose(1, 5, strides=2, padding='same', use_bias=False, activation='tanh')(x)
    return models.Model([noise_input, label_input], output, name='generator')

# 5. CGAN discriminator architecture

def make_discriminator_model(num_classes=10):
    img_input = layers.Input(shape=(28, 28, 1))
    label_input = layers.Input(shape=(1,), dtype='int32')
    label_embedding = layers.Embedding(num_classes, 28*28, input_length=1)(label_input)
    label_embedding = layers.Flatten()(label_embedding)
    label_reshape = layers.Reshape((28, 28, 1))(label_embedding)
    x = layers.Concatenate()([img_input, label_reshape])
    x = layers.Conv2D(64, 5, strides=2, padding='same')(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Conv2D(128, 5, strides=2, padding='same')(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Flatten()(x)
    output = layers.Dense(1)(x)
    return models.Model([img_input, label_input], output, name='discriminator')

# 6. CGAN model with binary crossentropy loss
class CGAN(models.Model):
    def __init__(self, D, G, latent_dim):
        super().__init__()
        self.D = D
        self.G = G
        self.latent_dim = latent_dim
        self.loss_fn = tf.keras.losses.BinaryCrossentropy(from_logits=True)
        self.d_optimizer = None
        self.g_optimizer = None
        self.d_loss_metric = metrics.Mean(name="d_loss")
        self.g_loss_metric = metrics.Mean(name="g_loss")

    @property
    def metrics(self):
        return [self.d_loss_metric, self.g_loss_metric]

    def compile(self, d_optimizer, g_optimizer):
        super().compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer

    def train_step(self, data):
        real_images, real_labels = data
        batch_size = tf.shape(real_images)[0]
        random_lat = tf.random.normal((batch_size, self.latent_dim))
        random_labels = tf.random.uniform((batch_size, 1), minval=0, maxval=num_classes, dtype='int32')

        with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
            fake_images = self.G([random_lat, random_labels], training=True)
            real_logits = self.D([real_images, tf.expand_dims(real_labels, 1)], training=True)
            fake_logits = self.D([fake_images, random_labels], training=True)
            real_labels_smooth = tf.ones_like(real_logits) - 0.1 * tf.random.uniform(tf.shape(real_logits))
            fake_labels_smooth = tf.zeros_like(fake_logits) + 0.1 * tf.random.uniform(tf.shape(fake_logits))
            d_loss_real = self.loss_fn(real_labels_smooth, real_logits)
            d_loss_fake = self.loss_fn(fake_labels_smooth, fake_logits)
            d_loss = 0.5 * (d_loss_real + d_loss_fake)
            g_loss = self.loss_fn(tf.ones_like(fake_logits), fake_logits)

        grads_d = disc_tape.gradient(d_loss, self.D.trainable_variables)
        grads_g = gen_tape.gradient(g_loss, self.G.trainable_variables)
        self.d_optimizer.apply_gradients(zip(grads_d, self.D.trainable_variables))
        self.g_optimizer.apply_gradients(zip(grads_g, self.G.trainable_variables))

        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)
        return {"d_loss": self.d_loss_metric.result(), "g_loss": self.g_loss_metric.result()}

# 7. Callback to save class-conditional samples every 10 epochs
class SaveResultsCallback(Callback):
    def __init__(self, output_dir, latent_dim, num_classes=10):
        super().__init__()
        self.output_dir = output_dir
        self.latent_dim = latent_dim
        self.num_classes = num_classes
        self.epochs = []
        self.d_losses = []
        self.g_losses = []

    def on_epoch_end(self, epoch, logs=None):
        e = epoch + 1
        d = logs["d_loss"]
        g = logs["g_loss"]
        print(f"Epoch {e}/{self.params['epochs']} – d_loss:{d:.4f} – g_loss:{g:.4f}")

        self.epochs.append(e)
        self.d_losses.append(d)
        self.g_losses.append(g)

        if e % 10 == 0:
            z = tf.random.normal((self.num_classes, self.latent_dim))
            labels = tf.range(self.num_classes, dtype='int32')[:, None]
            imgs = self.model.G([z, labels], training=False)
            imgs = (imgs * 127.5 + 127.5).numpy().astype('uint8')

            fig = plt.figure(figsize=(self.num_classes*2, 2))
            for i, img in enumerate(imgs):
                ax = fig.add_subplot(1, self.num_classes, i+1)
                ax.imshow(img.squeeze(), cmap='gray')
                ax.set_title(str(i))
                ax.axis('off')
            fig.suptitle(f"Epoch {e} - Class Samples")
            sample_path = os.path.join(self.output_dir, f"class_samples_epoch_{e}.png")
            fig.savefig(sample_path)
            plt.show()
            plt.close(fig)
            print(f"→ Saved class-conditional samples to {sample_path}")

            fig2 = plt.figure()
            plt.plot(self.epochs, self.d_losses, label='d_loss')
            plt.plot(self.epochs, self.g_losses, label='g_loss')
            plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
            loss_path = os.path.join(self.output_dir, f"loss_epoch_{e}.png")
            fig2.savefig(loss_path)
            plt.show()
            plt.close(fig2)
            print(f"→ Saved loss curve to {loss_path}")

        if e % 100 == 0:
            wpath = os.path.join(self.output_dir, f"generator_epoch_{e}.weights.h5")
            self.model.G.save_weights(wpath)
            print(f"→ Saved generator weights to {wpath}")

# 8. Instantiate models, CGAN, optimizers, callbacks
G = make_generator_model(latent_dim, num_classes)
D = make_discriminator_model(num_classes)
cgan = CGAN(D, G, latent_dim)

d_optimizer = optimizers.Adam(2e-4, beta_1=0.5)
g_optimizer = optimizers.Adam(2e-4, beta_1=0.5)
cgan.compile(d_optimizer=d_optimizer, g_optimizer=g_optimizer)

reduce_lr = ReduceLROnPlateau(
    monitor='g_loss',
    factor=0.5,
    patience=10,
    min_lr=1e-5,
    verbose=1
)
saver = SaveResultsCallback(RUN_DIR, latent_dim, num_classes)

# 9. Train
EPOCHS = 2000
cgan.fit(
    train_ds,
    epochs=EPOCHS,
    callbacks=[reduce_lr, saver],
    verbose=1
)
```
