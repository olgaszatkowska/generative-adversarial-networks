```python
import tensorflow as tf
import os, numpy as np, matplotlib.pyplot as plt
from tensorflow.keras import layers, models, optimizers, metrics
from tensorflow.keras.callbacks import Callback, ReduceLROnPlateau
from google.colab import drive

print("TensorFlow version:", tf.__version__)
print("GPUs available:", tf.config.list_physical_devices('GPU'))

drive.mount('/content/drive', force_remount=True)

BASE_DIR = "/content/drive/MyDrive/DCGAN_results"
os.makedirs(BASE_DIR, exist_ok=True)
existing = sorted([d for d in os.listdir(BASE_DIR) if d.startswith("result")])
run_id = len(existing)            # pierwszy run → result0, następny → result1, …
RUN_DIR = os.path.join(BASE_DIR, f"result{run_id}")
os.makedirs(RUN_DIR, exist_ok=True)
print("Saving this run to:", RUN_DIR)

from tensorflow.keras.datasets import fashion_mnist
(train_images, _), _ = fashion_mnist.load_data()
train_images = train_images.reshape(-1, 28, 28, 1).astype('float32')
train_images = (train_images - 127.5) / 127.5  # Normalize to [-1, 1]

BUFFER_SIZE = 60000
BATCH_SIZE  = 256
train_ds = (
    tf.data.Dataset
      .from_tensor_slices(train_images)
      .shuffle(BUFFER_SIZE)
      .batch(BATCH_SIZE, drop_remainder=True)
      .prefetch(tf.data.AUTOTUNE)
)

def make_generator_model(latent_dim=100):
    model = models.Sequential(name='generator')
    model.add(layers.Dense(7*7*256, use_bias=False, input_shape=(latent_dim,)))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU(0.2))
    model.add(layers.Reshape((7, 7, 256)))
    model.add(layers.Conv2DTranspose(128, 5, strides=1, padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU(0.2))
    model.add(layers.Conv2DTranspose(64, 5, strides=2, padding='same', use_bias=False))
    model.add(layers.BatchNormalization())
    model.add(layers.LeakyReLU(0.2))
    model.add(layers.Conv2DTranspose(1, 5, strides=2, padding='same', use_bias=False, activation='tanh'))
    return model

def make_discriminator_model():
    model = models.Sequential(name='discriminator')
    model.add(layers.Conv2D(64, 5, strides=2, padding='same', input_shape=[28, 28, 1]))
    model.add(layers.LeakyReLU(0.2))
    model.add(layers.Dropout(0.3))
    model.add(layers.Conv2D(128, 5, strides=2, padding='same'))
    model.add(layers.LeakyReLU(0.2))
    model.add(layers.Dropout(0.3))
    model.add(layers.Flatten())
    model.add(layers.Dense(1))  # logits
    return model

class DCGAN(models.Model):
    def __init__(self, D, G, latent_dim):
        super().__init__()
        self.D = D
        self.G = G
        self.latent_dim = latent_dim
        self.loss_fn = tf.keras.losses.BinaryCrossentropy(from_logits=True)
        self.d_optimizer = None
        self.g_optimizer = None

        # metrics
        self.d_loss_metric = metrics.Mean(name="d_loss")
        self.g_loss_metric = metrics.Mean(name="g_loss")

    @property
    def metrics(self):
        return [self.d_loss_metric, self.g_loss_metric]

    def compile(self, d_optimizer, g_optimizer):
        super().compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer

    def train_step(self, real_images):
        batch_size = tf.shape(real_images)[0]
        random_lat = tf.random.normal((batch_size, self.latent_dim))

        # forward + loss
        with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
            fake_images = self.G(random_lat, training=True)

            real_logits = self.D(real_images, training=True)
            fake_logits = self.D(fake_images, training=True)

            real_labels = tf.ones_like(real_logits) - 0.1 * tf.random.uniform(tf.shape(real_logits))
            fake_labels = tf.zeros_like(fake_logits) + 0.1 * tf.random.uniform(tf.shape(fake_logits))

            d_loss_real = self.loss_fn(real_labels, real_logits)
            d_loss_fake = self.loss_fn(fake_labels, fake_logits)
            d_loss = 0.5 * (d_loss_real + d_loss_fake)
            g_loss = self.loss_fn(tf.ones_like(fake_logits), fake_logits)

        # backward + update
        grads_d = disc_tape.gradient(d_loss, self.D.trainable_variables)
        grads_g = gen_tape.gradient(g_loss, self.G.trainable_variables)
        self.d_optimizer.apply_gradients(zip(grads_d, self.D.trainable_variables))
        self.g_optimizer.apply_gradients(zip(grads_g, self.G.trainable_variables))

        # update metrics
        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)
        return {"d_loss": self.d_loss_metric.result(), "g_loss": self.g_loss_metric.result()}

class SaveResultsCallback(Callback):
    def __init__(self, output_dir, latent_dim, num_images=5):
        super().__init__()
        self.output_dir = output_dir
        self.latent_dim = latent_dim
        self.num_images = num_images
        self.epochs = []
        self.d_losses = []
        self.g_losses = []

    def on_epoch_end(self, epoch, logs=None):
        e = epoch + 1
        d = logs["d_loss"]
        g = logs["g_loss"]
        print(f"Epoch {e}/{self.params['epochs']} – d_loss:{d:.4f} – g_loss:{g:.4f}")

        self.epochs.append(e)
        self.d_losses.append(d)
        self.g_losses.append(g)

        # co 10 epok: zapisz próbki i wykresy strat
        if e % 10 == 0:
            # -- samples --
            z = tf.random.normal((self.num_images, self.latent_dim))
            imgs = self.model.G(z, training=False)
            imgs = (imgs * 127.5 + 127.5).numpy().astype('uint8')

            fig = plt.figure(figsize=(self.num_images*2, 2))
            for i, img in enumerate(imgs):
                ax = fig.add_subplot(1, self.num_images, i+1)
                ax.imshow(img.squeeze(), cmap='gray')
                ax.axis('off')
            fig.suptitle(f"Epoch {e}")
            sample_path = os.path.join(self.output_dir, f"samples_epoch_{e}.png")
            fig.savefig(sample_path)
            plt.show()
            plt.close(fig)
            print(f"→ Saved samples to {sample_path}")

            # -- loss curve --
            fig2 = plt.figure()
            plt.plot(self.epochs, self.d_losses, label='d_loss')
            plt.plot(self.epochs, self.g_losses, label='g_loss')
            plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
            loss_path = os.path.join(self.output_dir, f"loss_epoch_{e}.png")
            fig2.savefig(loss_path)
            plt.show()
            plt.close(fig2)
            print(f"→ Saved loss curve to {loss_path}")

        # co 100 epok: zapisz wagi generatora
        if e % 100 == 0:
            # Correct the file extension to .weights.h5
            wpath = os.path.join(self.output_dir, f"generator_epoch_{e}.weights.h5")
            self.model.G.save_weights(wpath)
            print(f"→ Saved generator weights to {wpath}")

latent_dim = 100
G = make_generator_model(latent_dim)
D = make_discriminator_model()
dcgan = DCGAN(D, G, latent_dim)

d_optimizer = optimizers.Adam(2e-4, beta_1=0.5)
g_optimizer = optimizers.Adam(2e-4, beta_1=0.5)
dcgan.compile(d_optimizer=d_optimizer, g_optimizer=g_optimizer)

reduce_lr = ReduceLROnPlateau(
    monitor='g_loss',
    factor=0.5,
    patience=10,
    min_lr=1e-5,
    verbose=1
)
saver = SaveResultsCallback(RUN_DIR, latent_dim, num_images=5)

EPOCHS = 2000
dcgan.fit(
    train_ds,
    epochs=EPOCHS,
    callbacks=[reduce_lr, saver],
    verbose=1
)
```
