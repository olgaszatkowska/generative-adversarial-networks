```python
import tensorflow as tf
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models, optimizers, metrics
from tensorflow.keras.callbacks import Callback, ReduceLROnPlateau
from google.colab import drive

print("TensorFlow version:", tf.__version__)
print("GPUs available:", tf.config.list_physical_devices('GPU'))

drive.mount('/content/drive', force_remount=True)

BASE_DIR = "/content/drive/MyDrive/CGAN_results"
os.makedirs(BASE_DIR, exist_ok=True)
existing = sorted([d for d in os.listdir(BASE_DIR) if d.startswith("result")])
run_id = len(existing)
RUN_DIR = os.path.join(BASE_DIR, f"result{run_id}")
os.makedirs(RUN_DIR, exist_ok=True)
print("Saving this run to:", RUN_DIR)

(train_images, train_labels), _ = tf.keras.datasets.mnist.load_data()
train_images = train_images.reshape(-1, 28, 28, 1).astype('float32')
train_images = (train_images - 127.5) / 127.5  # Normalize to [-1, 1]

num_classes = 10
train_labels = tf.one_hot(train_labels, depth=num_classes)

BUFFER_SIZE = 60000
BATCH_SIZE = 256
train_ds = (
    tf.data.Dataset.from_tensor_slices((train_images, train_labels))
      .shuffle(BUFFER_SIZE)
      .batch(BATCH_SIZE, drop_remainder=True)
      .prefetch(tf.data.AUTOTUNE)
)

def make_generator_model(latent_dim=100, num_classes=10):
    z_input = layers.Input(shape=(latent_dim,), name='z_input')
    label_input = layers.Input(shape=(num_classes,), name='label_input')
    # concatenate noise and label
    x = layers.Concatenate()([z_input, label_input])
    x = layers.Dense(7*7*256, use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Reshape((7, 7, 256))(x)

    x = layers.Conv2DTranspose(128, 5, strides=1, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)

    x = layers.Conv2DTranspose(64, 5, strides=2, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.LeakyReLU(0.2)(x)

    img_output = layers.Conv2DTranspose(1, 5, strides=2, padding='same', use_bias=False, activation='tanh')(x)
    return models.Model([z_input, label_input], img_output, name='c_generator')

def make_discriminator_model(num_classes=10):
    img_input = layers.Input(shape=(28, 28, 1), name='img_input')
    label_input = layers.Input(shape=(num_classes,), name='label_input')
    # expand label to image size and concat
    y = layers.Dense(28*28, activation='relu')(label_input)
    y = layers.Reshape((28, 28, 1))(y)
    x = layers.Concatenate(axis=-1)([img_input, y])

    x = layers.Conv2D(64, 5, strides=2, padding='same')(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Conv2D(128, 5, strides=2, padding='same')(x)
    x = layers.LeakyReLU(0.2)(x)
    x = layers.Dropout(0.3)(x)

    x = layers.Flatten()(x)
    validity = layers.Dense(1)(x)  # logits
    return models.Model([img_input, label_input], validity, name='c_discriminator')

class CGAN(models.Model):
    def __init__(self, discriminator, generator, latent_dim, num_classes):
        super().__init__()
        self.discriminator = discriminator
        self.generator = generator
        self.latent_dim = latent_dim
        self.num_classes = num_classes
        self.loss_fn = tf.keras.losses.BinaryCrossentropy(from_logits=True)
        self.d_loss_metric = metrics.Mean(name="d_loss")
        self.g_loss_metric = metrics.Mean(name="g_loss")

    @property
    def metrics(self):
        return [self.d_loss_metric, self.g_loss_metric]

    def compile(self, d_optimizer, g_optimizer):
        super().compile()
        self.d_optimizer = d_optimizer
        self.g_optimizer = g_optimizer

    def train_step(self, data):
        real_images, real_labels = data
        batch_size = tf.shape(real_images)[0]
        # sample noise and random labels
        random_lat = tf.random.normal((batch_size, self.latent_dim))
        random_labels = tf.one_hot(
            tf.random.uniform((batch_size,), minval=0, maxval=self.num_classes, dtype=tf.int32),
            depth=self.num_classes
        )

        with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
            fake_images = self.generator([random_lat, random_labels], training=True)

            real_logits = self.discriminator([real_images, real_labels], training=True)
            fake_logits = self.discriminator([fake_images, random_labels], training=True)

            real_targets = tf.ones_like(real_logits) - 0.1 * tf.random.uniform(tf.shape(real_logits))
            fake_targets = tf.zeros_like(fake_logits) + 0.1 * tf.random.uniform(tf.shape(fake_logits))

            d_loss_real = self.loss_fn(real_targets, real_logits)
            d_loss_fake = self.loss_fn(fake_targets, fake_logits)
            d_loss = 0.5 * (d_loss_real + d_loss_fake)
            g_loss = self.loss_fn(tf.ones_like(fake_logits), fake_logits)

        grads_d = disc_tape.gradient(d_loss, self.discriminator.trainable_variables)
        grads_g = gen_tape.gradient(g_loss, self.generator.trainable_variables)
        self.d_optimizer.apply_gradients(zip(grads_d, self.discriminator.trainable_variables))
        self.g_optimizer.apply_gradients(zip(grads_g, self.generator.trainable_variables))

        self.d_loss_metric.update_state(d_loss)
        self.g_loss_metric.update_state(g_loss)
        return {"d_loss": self.d_loss_metric.result(), "g_loss": self.g_loss_metric.result()}

class SaveResultsCallback(Callback):
    def __init__(self, output_dir, latent_dim, num_classes, num_images=10):
        super().__init__()
        self.output_dir = output_dir
        self.latent_dim = latent_dim
        self.num_classes = num_classes
        self.num_images = num_images
        self.epochs = []
        self.d_losses = []
        self.g_losses = []
        # fixed labels 0..num_classes-1
        self.fixed_labels = tf.one_hot(np.arange(num_classes), depth=num_classes)

    def on_epoch_end(self, epoch, logs=None):
        e = epoch + 1
        # It's good practice to check if logs are available and contain the keys
        d = logs.get("d_loss") if logs else None
        g = logs.get("g_loss") if logs else None

        # Only print/append if logs are available
        if d is not None and g is not None:
            print(f"Epoch {e}/{self.params['epochs']} – d_loss:{d:.4f} – g_loss:{g:.4f}")

            self.epochs.append(e)
            self.d_losses.append(d)
            self.g_losses.append(g)
        else:
            print(f"Epoch {e}/{self.params['epochs']} – Logs not available")


        # co 10 epok: próbki i wykres strat
        if e % 10 == 0 and logs is not None:
            z = tf.random.normal((self.num_images, self.latent_dim))
            imgs = self.model.generator([z, self.fixed_labels], training=False)
            imgs = (imgs * 127.5 + 127.5).numpy().astype('uint8')

            fig = plt.figure(figsize=(self.num_images*2, 2))
            for i, img in enumerate(imgs):
                ax = fig.add_subplot(1, self.num_images, i+1)
                ax.imshow(img.squeeze(), cmap='gray')
                ax.axis('off')
            fig.suptitle(f"Epoch {e}")
            sample_path = os.path.join(self.output_dir, f"samples_epoch_{e}.png")
            fig.savefig(sample_path)
            plt.show()
            plt.close(fig)
            print(f"→ Saved samples to {sample_path}")

            # Plot loss only if we have accumulated data
            if self.epochs:
                fig2 = plt.figure()
                plt.plot(self.epochs, self.d_losses, label='d_loss')
                plt.plot(self.epochs, self.g_losses, label='g_loss')
                plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
                loss_path = os.path.join(self.output_dir, f"loss_epoch_{e}.png")
                fig2.savefig(loss_path)
                plt.show()
                plt.close(fig2)
                print(f"→ Saved loss curve to {loss_path}")


        # co 100 epok: wagi generatora
        if e % 100 == 0:
            # Change the file extension to .weights.h5 as required by TensorFlow
            wpath = os.path.join(self.output_dir, f"generator_epoch_{e}.weights.h5")
            self.model.generator.save_weights(wpath)
            print(f"→ Saved generator weights to {wpath}")

# 8. Inicjalizacja modeli, CGAN, optymalizatory, callbacki
latent_dim = 100
num_classes = 10
G = make_generator_model(latent_dim, num_classes)
D = make_discriminator_model(num_classes)
cgan = CGAN(D, G, latent_dim, num_classes)

d_optimizer = optimizers.Adam(2e-4, beta_1=0.5)
g_optimizer = optimizers.Adam(2e-4, beta_1=0.5)
cgan.compile(d_optimizer=d_optimizer, g_optimizer=g_optimizer)

reduce_lr = ReduceLROnPlateau(
    monitor='g_loss',
    factor=0.5,
    patience=10,
    min_lr=1e-5,
    verbose=1
)
saver = SaveResultsCallback(RUN_DIR, latent_dim, num_classes, num_images=num_classes)

epochs = 2000
cgan.fit(
    train_ds,
    epochs=epochs,
    callbacks=[reduce_lr, saver],
    verbose=1
)
```
